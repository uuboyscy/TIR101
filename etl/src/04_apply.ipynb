```python
def reserve_int_value(input_str: str) -> int:
    """
    Convert string to int.

    Example:
        "推: 25" -> 25
    """
    return int(input_str.split(" ")[1])
```

```python
df["推"].apply(reserve_int_value)
```

```python
# df["推"] = df["推"].apply(reserve_int_value)
df["推_int"] = df["推"].apply(reserve_int_value)
```

```python
from pandas._libs.tslibs import Timestamp

def convert_time(input_datetime_string: str) -> Timestamp:
    datetime_str = input_datetime_string.replace("時間: ", "")
    return pd.to_datetime(datetime_str)
```

```python
df["時間"].apply(convert_time)
```

```python
df["時間_dt"] = df["時間"].apply(convert_time)
df.head()
```
