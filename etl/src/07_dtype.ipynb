```python
column_dtypes = {
    "DepartmentID": str,
    "Intern": bool,
    "CreatedDatetime": "datetime64[ns]"
}

df = pd.read_csv(
    file_path,
    dtype=column_dtypes,
)

df
```

```python
column_dtypes = {
    "DepartmentID": str,
    "Intern": bool,
}

df = pd.read_csv(
    file_path,
    dtype=column_dtypes,
)

df
```

```python
df["Experience"] = np.where(
    df["Experience"].isna(),
    0,
    df["Experience"],
)

df
```

```python
df["OnBoard"] = np.where(
    df["OnBoard"].str.strip() == "V",
    True,
    False,
)

df
```

```python
pd.to_datetime(df["CreatedDatetime"])
```

```python
pd.to_datetime(df["CreatedDatetimeTZ"])
```

```python
column_dtypes = {
    "DepartmentID": str,
    "Intern": bool,
}

df = pd.read_csv(
    file_path,
    dtype=column_dtypes,
    parse_dates=["CreatedDatetime", "CreatedDatetimeTZ"],
    date_parser=lambda x: pd.to_datetime(x)
)

df
```
